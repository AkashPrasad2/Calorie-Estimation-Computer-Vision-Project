# Image Processing and Preparation

```python
import os
import json
from PIL import Image, ImageDraw
import torch
from torch.utils.data import Dataset
import numpy as np
import random

class BananaSegmentationDataset(Dataset):
    def __init__(self, paired_files, transform=None, mask_transform=None):
        self.paired_files = paired_files
        self.transform = transform
        self.mask_transform = mask_transform

    def __len__(self):
        return len(self.paired_files)

    def __getitem__(self, idx):
        image_path, mask_path = self.paired_files[idx]

        # Load the image
        image = Image.open(image_path).convert('RGB')

        # Create the mask
        mask = self.create_mask_from_json(mask_path, image.size)

        # Apply transforms
        if self.transform:
            image = self.transform(image)
        if self.mask_transform:
            mask = self.mask_transform(mask)

        return image, mask

    def create_mask_from_json(self, json_path, image_size):
        # Create a blank mask with all zeros (background)
        mask = Image.new('L', image_size, 0)
        draw = ImageDraw.Draw(mask)

        with open(json_path, 'r') as f:
            data = json.load(f)

            # Draw polygons for each class
            for shape in data['shapes']:
                label = shape['label'].lower()
                points = shape['points']
                polygon = [(int(point[0]), int(point[1])) for point in points]

                if label == 'banana':
                    draw.polygon(polygon, outline=1, fill=1)  # Class 1: Banana
                elif label == 'thumb':
                    draw.polygon(polygon, outline=2, fill=2)  # Class 2: Thumb
                elif label == 'apple':
                    draw.polygon(polygon, outline=3, fill=3)  # Class 3: Apple
                # elif label == 'bread':
                #     draw.polygon(polygon, outline=4, fill=4)  # Class 4: bread

        return mask

```

## Fix name parining issue

```python
def pair_images_and_masks(image_dir, mask_dir):
    # List all image and mask files
    image_files = sorted([f for f in os.listdir(image_dir) if f.lower().endswith('.jpg')])
    mask_files = sorted([f for f in os.listdir(mask_dir) if f.endswith('.json')])

    # Pair images and masks by matching base names
    paired_files = []
    for img in image_files:
        base_name = os.path.splitext(img)[0]  # Extract base name (e.g., '1 (1)')
        corresponding_mask = f"{base_name}.json"  # Mask file should have the same base name

        if corresponding_mask in mask_files:
            paired_files.append((os.path.join(image_dir, img), os.path.join(mask_dir, corresponding_mask)))
        else:
            print(f"Warning: No mask found for image {img}")

    return paired_files
```

# Building the Neural Network

## Attention Mechanism

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class CBAM(nn.Module):
    def __init__(self, in_channels, reduction_ratio=16):
        super(CBAM, self).__init__()

        # Channel Attention
        self.channel_avg_pool = nn.AdaptiveAvgPool2d(1)
        self.channel_max_pool = nn.AdaptiveMaxPool2d(1)
        self.shared_mlp = nn.Sequential(
            nn.Conv2d(in_channels, in_channels // reduction_ratio, kernel_size=1, bias=False),
            nn.ReLU(),
            nn.Conv2d(in_channels // reduction_ratio, in_channels, kernel_size=1, bias=False)
        )
        self.sigmoid_channel = nn.Sigmoid()

        # Spatial Attention
        self.spatial_conv = nn.Conv2d(2, 1, kernel_size=7, stride=1, padding=3, bias=False)
        self.sigmoid_spatial = nn.Sigmoid()

    def forward(self, x):
        # Channel Attention
        avg_out = self.channel_avg_pool(x)
        max_out = self.channel_max_pool(x)
        channel_attention = self.shared_mlp(avg_out) + self.shared_mlp(max_out)
        channel_attention = self.sigmoid_channel(channel_attention)
        x = x * channel_attention

        # Spatial Attention
        avg_out = torch.mean(x, dim=1, keepdim=True)
        max_out, _ = torch.max(x, dim=1, keepdim=True)
        spatial_attention = torch.cat([avg_out, max_out], dim=1)
        spatial_attention = self.spatial_conv(spatial_attention)
        spatial_attention = self.sigmoid_spatial(spatial_attention)
        x = x * spatial_attention

        return x
```

```python
import torch
import torch.nn as nn
import torchvision.models as models
import torch.nn.functional as F

class BananaSegmentationModelUpdated(nn.Module):
    def __init__(self):
        super(BananaSegmentationModelUpdated, self).__init__()

        # Initialize ResNet34
        resnet = models.resnet34(pretrained=True)
        self.encoder_conv1 = resnet.conv1  # Output: 64 channels
        self.bn1 = resnet.bn1
        self.relu = resnet.relu
        self.maxpool = resnet.maxpool  # Downsampled spatially

        self.layer1 = resnet.layer1  # Output: 64 channels
        self.layer2 = resnet.layer2  # Output: 128 channels
        self.layer3 = resnet.layer3  # Output: 256 channels
        self.layer4 = resnet.layer4  # Output: 512 channels

        # Add CBAM blocks with correct channel sizes
        self.cbam1 = CBAM(64)
        self.cbam2 = CBAM(128)
        self.cbam3 = CBAM(256)
        self.cbam4 = CBAM(512)

        # Decoder layers with unique names
        self.up1 = nn.ConvTranspose2d(512, 256, kernel_size=2, stride=2)
        self.decoder_conv1 = nn.Conv2d(512, 256, kernel_size=3, padding=1)
        self.up2 = nn.ConvTranspose2d(256, 128, kernel_size=2, stride=2)
        self.decoder_conv2 = nn.Conv2d(256, 128, kernel_size=3, padding=1)
        self.up3 = nn.ConvTranspose2d(128, 64, kernel_size=2, stride=2)
        self.decoder_conv3 = nn.Conv2d(128, 64, kernel_size=3, padding=1)
        self.up4 = nn.ConvTranspose2d(64, 64, kernel_size=2, stride=2)
        self.decoder_conv4 = nn.Conv2d(128, 64, kernel_size=3, padding=1)
        self.final_conv = nn.Conv2d(64, 4, kernel_size=1)

    def forward(self, x):
        # Encoder path with CBAM
        e1 = self.encoder_conv1(x)  # Input convolution
        e1 = self.bn1(e1)           # Batch normalization
        e1 = self.relu(e1)          # ReLU activation
        e1 = self.maxpool(e1)       # Max pooling
        e1 = self.cbam1(e1)         # Apply CBAM for 64 channels

        e2 = self.layer1(e1)        # ResNet layer 1
        e2 = self.cbam2(e2)         # Apply CBAM for 128 channels

        e3 = self.layer2(e2)        # ResNet layer 2
        e3 = self.cbam3(e3)         # Apply CBAM for 256 channels

        e4 = self.layer3(e3)        # ResNet layer 3
        e4 = self.cbam4(e4)         # Apply CBAM for 512 channels

        # Decoder path
        d4 = self.up1(e4)
        d4 = torch.cat([d4, e4], dim=1)
        d4 = self.decoder_conv1(d4)

        d3 = self.up2(d4)
        d3 = torch.cat([d3, e3], dim=1)
        d3 = self.decoder_conv2(d3)

        d2 = self.up3(d3)
        d2 = torch.cat([d2, e2], dim=1)
        d2 = self.decoder_conv3(d2)

        d1 = self.up4(d2)
        d1 = torch.cat([d1, e1], dim=1)
        d1 = self.decoder_conv4(d1)

        # Final output
        output = self.final_conv(d1)
        outputs = F.interpolate(output, size=(128, 128), mode='bilinear', align_corners=False)

        return outputs
```

# Training the Model

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split
import torchvision.transforms as transforms
# Directory paths
images_dir = r"/content/dataset/Images"
masks_dir = r"/content/dataset/masks"

paired_files = pair_images_and_masks(images_dir, masks_dir) # pair the image names properly

# Define transformations
image_transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.ToTensor(),
])

# Transformations for masks
mask_transform = transforms.Compose([
    transforms.Resize((128, 128), interpolation=Image.NEAREST),
    transforms.Lambda(lambda img: torch.from_numpy(np.array(img)).long())
])


# Initialize dataset and dataloaders
dataset = BananaSegmentationDataset(
    paired_files,
    transform=image_transform,
    mask_transform=mask_transform
)

# Split dataset into training and validation
train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size
train_dataset, val_dataset = random_split(dataset, [train_size, val_size])

# DataLoaders
train_loader = DataLoader(train_dataset, batch_size=1, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=1, shuffle=False)


# Model, optimizer, and loss function
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = BananaSegmentationModelUpdated().to(device)
optimizer = optim.Adam(model.parameters(), lr=1e-3)
criterion = nn.CrossEntropyLoss()

# Training function
def train_model(num_epochs=20):
    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        for images, masks in train_loader:
            images, masks = images.to(device), masks.to(device)

            outputs = model(images)  # Output shape: [batch_size, num_classes, height, width]
            masks = masks.squeeze(1).long()  # Remove channel dimension and convert to Long tensor

            loss = criterion(outputs, masks)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * images.size(0)


        epoch_loss = running_loss / len(train_loader.dataset)
        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {epoch_loss:.4f}")

train_model()
```

# Estimating Calories for pictures from the dataset
1. Calculate the number of pixels in the thumb manually using the known dimensions of it. This gives us a constant.
2. Calculate the pixel ratio between the thumb and the banana.
3. The ratio of pixels will be the same as the ratio of centimeters so use that to get the surface area of the banana in centimeters.
4. Assume a constant depth for the banana and calculate the volume by $Volume = SA* Depth$
5. We know the density of a banana: $1.15\frac{g}{cm^3}$
6. Rearranging the formula for density we get $g = Density*Volume$ and we know both of these values
7. Finally this gives us the number of grams in the banana and we know the number of calories per gram so we can easliy convert it.


Dimensions of the thumb are:
$5cm$ by $1.2cm$

```python
thumb_known_sa = 6.5 * 2  # Surface area of thumb in cm² (known dimensions)

# banana constants
banana_density = 1.15  # g/cm³
banana_cal_per_gram = 0.8  # Calories per gram of banana
banana_assumed_depth = 3  # Assumed depth of the banana in cm

# apple constants
apple_density = 0.75
apple_cal_per_gram = 0.52
apple_assumed_depth = 7.2

#bread constants
#bread_density =

def calculate_pixel_ratio(segmentation_mask, thumb_class, fruit_class):
    """
    Calculate the pixel ratio between the thumb and the fruit from the segmentation mask.

    Args:
        segmentation_mask (torch.Tensor): The segmentation mask output from the model.
        thumb_class (int): The class index for the thumb in the segmentation mask.
        fruit_class (int): The class index for the fruit in the segmentation mask.

    Returns:
        float: The pixel ratio of thumb to fruit.
    """
    thumb_pixels = torch.sum(segmentation_mask == thumb_class).item()
    fruit_pixels = torch.sum(segmentation_mask == fruit_class).item()

    if fruit_pixels == 0:
        return float('inf')  # Avoid division by zero
    else:
        pixel_ratio = thumb_pixels / fruit_pixels

    return pixel_ratio

def calculate_calories(segmentation_mask, thumb_class, fruit_class, density, cal_per_gram, assumed_depth):
    """
    Estimate the number of calories in a fruit using the segmentation mask.

    Args:
        segmentation_mask (torch.Tensor): The segmentation mask output from the model.
        thumb_class (int): The class index for the thumb in the segmentation mask.
        fruit_class (int): The class index for the fruit in the segmentation mask.
        density (float): The density of the fruit in g/cm³.
        cal_per_gram (float): Calories per gram of the fruit.
        assumed_depth (float): Assumed depth of the fruit in cm.

    Returns:
        float: Estimated number of calories in the fruit.
    """
    # Step 1: Calculate the pixel ratio
    pixel_ratio = calculate_pixel_ratio(segmentation_mask, thumb_class, fruit_class)

    if pixel_ratio == float('inf'):
        print(f"No fruit of class {fruit_class} detected in the segmentation mask.")
        return 0

    # Step 2: Calculate the surface area of the fruit in cm²
    fruit_sa = thumb_known_sa / pixel_ratio

    # Step 3: Calculate the volume of the fruit in cm³
    fruit_volume = fruit_sa * assumed_depth

    # Step 4: Calculate the mass of the fruit in grams
    fruit_mass = density * fruit_volume

    # Step 5: Calculate the calories of the fruit
    fruit_calories = fruit_mass * cal_per_gram

    return fruit_calories

# Example usage
thumb_class = 2  # Class index for the thumb
banana_class = 1  # Class index for the banana
apple_class = 3   # Class index for the apple

# Calculate calories for banana
banana_calories = calculate_calories(
    predicted_mask, thumb_class, banana_class,
    banana_density, banana_cal_per_gram, banana_assumed_depth
)
print(f"Estimated calories in the banana: {banana_calories:.2f} kcal")

# Calculate calories for apple
apple_calories = calculate_calories(
    predicted_mask, thumb_class, apple_class,
    apple_density, apple_cal_per_gram, apple_assumed_depth
)
print(f"Estimated calories in the apple: {apple_calories:.2f} kcal")
```
